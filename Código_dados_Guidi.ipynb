```python
def dados(df, nome_do_arquivo):

    '''
    Função que extrai os dados de uma página de um ponto especificado pela variável nome_do_arquivo
    df é um dataframe lido da mesma página

    #------------------------------------------x--------------------------------------------------#

    A variável df1 é onde os dados vão ser retornados pela função
    df2 é uma cópia de segurança

    '''
    
    for i in range(len(df.columns)):
        df[i] = df[i].astype(str) #df1 é a variavel (dataframe) em que os dados estão sendo inseridos
        
    df1 = df[0]       
    df2 = pd.DataFrame()
    df_copia = df.copy()
    
    for i in range(len(df.columns)):
        df[i] = df[i].astype(str)
            
    for i in range(1, len(df.columns) -1):
        if df[i+1].str.contains("^±").any():
            df1 = pd.concat([df1, df[i]], axis = 1)
            break
            
        elif df[i].str.contains("±.[0-9]{1,5}$").any():
            df1 = pd.concat([df1, df[i].str.extract("(^[0-9]{1,7}.)")], axis = 1)
            break
    
    df1.columns = ["Element", "Peak Data"]
    df1 = pd.concat([df1, df1["Element"].str.extract("([A-Za-z]{1,3}.[A-Z]$)")], axis = 1)
    df1.columns = ["Lines", "Peak Data", "Element"]
    
    df1 = df1.dropna()

    df1["Nome_do_Arquivo"] = nome_do_arquivo
    df1.set_index("Element", inplace = True)  # <-- append = True
    
    return df1


        
```

```python
def pagina(nome_do_arquivo, num_pag):
    '''
    Função que junta todas as páginas utilizando a função Dados
    '''
    try:
        for i in range(0, num_pag): 
            if i == 0:
                df = dados(tabula.read_pdf(nome_do_arquivo, pages = "all",  pandas_options = {"header": None}, multiple_tables = True)[0][3:], nome_do_arquivo)
        
            else:
                df = pd.concat([df, dados(tabula.read_pdf(nome_do_arquivo, pages = "all",  pandas_options = {"header": None}, multiple_tables = True)[i][:], nome_do_arquivo)])
            
    except:
        for i in range(0, num_pag-1): 
            if i == 0:
                df = dados(tabula.read_pdf(nome_do_arquivo, pages = "all",  pandas_options = {"header": None}, multiple_tables = True)[0][3:], nome_do_arquivo)
        
            else:
                df = pd.concat([df, dados(tabula.read_pdf(nome_do_arquivo, pages = "all",  pandas_options = {"header": None}, multiple_tables = True)[i][:], nome_do_arquivo)])
        
    df["Peak Data"] = df["Peak Data"].astype(float) 
    
    
    return df
```
