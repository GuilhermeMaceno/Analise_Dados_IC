```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Defina as categorias e os valores para cada categoria
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [4, 3, 2, 5, 4]

# Crie um array de valores para o ângulo theta
theta = np.linspace(0, 2*np.pi, len(categorias), endpoint=False)

# Feche o gráfico
valores += valores[:1]
theta = np.append(theta, theta[:1])

# Crie o gráfico
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(theta, valores, 'b', alpha=0.1)

# Adicione rótulos nas categorias
ax.set_xticks(theta[:-1])
ax.set_xticklabels(categorias)

plt.show()

```
